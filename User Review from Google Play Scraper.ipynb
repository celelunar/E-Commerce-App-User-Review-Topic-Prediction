```python
# Google Play Scraper
from google_play_scraper import app
from google_play_scraper import Sort, reviews

# Data Manipulation
import pandas as pd
import numpy as np
```

## Get the Reviews to A DataFrame

```python
# Dictionary to map app IDs to app names
app_id_to_name = {
    'com.shopee.id': 'Shopee',
    'com.tokopedia.tkpd': 'Tokopedia',
    'com.lazada.android': 'Lazada',
    'blibli.mobile.commerce': 'Blibli',
    'com.zalora.android':'Zalora',
    'com.bukalapak.android': 'Bukalapak',
    'com.app.tokobagus.betterb' : 'OLX',
    'com.alibaba.intl.android.apps.poseidon' : 'Alibaba'

}

# List of app IDs to scrape reviews from
app_ids = list(app_id_to_name.keys())

# List to hold all reviews
all_reviews = []

# Iterate through each app ID and get reviews
for app_id in app_ids:
    result, continuation_token = reviews(
        app_id,
        lang='eng',
        country='us',
        sort=Sort.MOST_RELEVANT,
        count=3000,
        filter_score_with=None
    )

    # Append the app name to each review in the result
    for review in result:
        review['app_name'] = app_id_to_name[app_id]

    all_reviews.extend(result)
```

```python
df_review = pd.DataFrame(np.array(all_reviews),columns=['review'])

df_review = df_review.join(pd.DataFrame(df_review.pop('review').tolist()))

df_review.head()
```

## Take Only The Important Columns

```python
df_review[['userName', 'score','at', 'content', 'app_name']].head()
```

## Sort the Reviews Based on The Newest

```python
new_df = df_review[['userName', 'score','at', 'content', 'app_name']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score','at', 'content', 'app_name']] #get userName, rating, date-time, reviews, and app name
```
